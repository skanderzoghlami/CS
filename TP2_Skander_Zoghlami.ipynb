#### Correction :
on peut remarquer la présence de la valeur -99 dans l'histogramme des weights et de heights qui peut étre due a un manque d'informations sur l'héro en question ,pour corriger ce probléme on peut  remplacer les valeurs par :
- la valeur mean : la moyenne des valeurs du dataset.
- La valeur median : la valeur mediane du dataset
- La valeur mode : a valeur la plus fréquante dans le dataset.
- ou on peut supprimer les lignes contenant cette valeur(cette option peut causer une perte de données qui peuvent étre importante si on veut appliquer un algorithme) 

* par exemple si on a la liste suivante [200 ,110 , 150 , 110 ,110]
- la valeur mean est :  (200 + 110 + 150 + 110 + 110)/5 = 136
- la valeur median est : on regroupe la liste et on fait un sort : [110 , 110 , 110 , 150 ,200] puis on choisie la valeur mediane : 110 
- la valeur mode est 110 puisque c'est le nombre le plus fréquant.
###### Dans notre cas on va appliquer la valeur moyenne

```python
import numpy as np
# On applique la fonction lambda sur les deux dataset et on verifie en affichant l'histogramme
df['Height'] = df['Height'].apply(lambda x : x if x > 0 else df['Height'].mean())
df['Weight'] = df['Weight'].apply(lambda x : x if x > 0 else df['Weight'].mean())
df['Weight'].hist()
```

## Question 6: (3 point)

Pour cette question vous êtes appelés à créer une nouvelle colonne IMC ( IMC = poids (kg) / taille² (m)).
* moins de 18,5 	Insuffisance pondérale (maigreur)
* 18,5 à 25	Corpulence normale
* 25 à 30	Surpoids
* 30 à 35	Obésité modérée
* 35 à 40	Obésité sévère
* plus de 40	Obésité morbide ou massive

Ensuite, visualisez à l'aide d'un histogramme le nombre de héros par catégorie d'IMC (les catégories doivent être libellées comme indiqué ci dessus)<br>
**NB: Faites attention pour les unités des colonnes et créez autant de colonnes que nécessaire**

```python
# Puisque le weight est en pounds on doit les convertir en KG
# puisque le height est en CM on doit le convertir en m
# 1 pound  = 0.45359237 Kgs
c = 0.45359237
def IMC(data):
    return (data['Weight']*c) / (data['Height']/100)**2
df['IMC_value'] = df.apply(IMC, axis=1)
def IMC_cat(data):
    if float(data['IMC_value'])<18.5:
        return "Insuffisance pondérale (maigreur)"
    elif (float(data['IMC_value'])<25.0) & (float(data['IMC_value'])>=18.5):
        return "Corpulence normale"
    elif (float(data['IMC_value'])<30.0) & (float(data['IMC_value'])>=25.0):
         return "Surpoids"
    elif (float(data['IMC_value'])<35.0) & (float(data['IMC_value'])>=30.0): 
        return "Obésité modérée"
    elif (float(data['IMC_value'])<40.0) & (float(data['IMC_value'])>=35.0): 
        return "Obésité sévère" 
    else:
        return "Obésité morbide ou massive"
df['IMC'] = df.apply(IMC_cat, axis=1)
```

```python
df.head()
```

```python
IMC1 = df.loc[((df["IMC"]=="Insuffisance pondérale (maigreur)"))]
IMC2 = df.loc[((df["IMC"]=="Corpulence normale)"))]
IMC3 = df.loc[((df["IMC"]=="Surpoids"))]
IMC4 = df.loc[((df["IMC"]=="Obésité modérée"))]
IMC5 = df.loc[((df["IMC"]=="Obésité sévère"))]
# Next we plot a histograms representation
plt.figure(figsize = (50,18))
plt.bar(["Insuffisance pondérale (maigreur)", "Corpulence normale", "Surpoids","Obésité modérée","Obésité sévère"] , [len(IMC1) , len(IMC2) , len(IMC3),len(IMC4),len(IMC5)] )
plt.xlabel("IMC category")
plt.ylabel("Number of heroes ")
plt.title("IMC bins")
```
